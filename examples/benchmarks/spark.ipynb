## Run ELTBench in `light` mode

```python
from lakebench.engines import Spark
from lakebench.benchmarks import ELTBench

engine = Spark(
    schema_name = 'spark_eltbench_test',
    schema_uri = 'C:/lakebench_local/working_dir/eltbench_sf1/',
    spark_measure_telemetry = False
)

benchmark = ELTBench(
    engine=engine,
    scenario_name="SF1",
    input_parquet_folder_uri='C:/lakebench_local/source/tpcds_sf1',
    save_results=True,
    result_table_uri='C:/lakebench_local/results'
    )
benchmark.run(mode="light")
```
