## Run ELTBench in `light` mode

```python
from lakebench.engines import Daft
from lakebench.benchmarks import ELTBench

engine = Daft(
    delta_abfss_schema_path = 'abfss://.........../Tables/daft_eltbench_test'
)

benchmark = ELTBench(
    engine=engine,
    scenario_name="SF1",
    tpcds_parquet_mount_path='/lakehouse/default/Files/tpcds_sf1',
    save_results=True,
    result_abfss_path='abfss://......../Tables/lakebench/results'
    )
benchmark.run(mode="light")
```
